```python
import glob
file_paths = glob.glob("*.csv")

dfs = []
for path in file_paths:
    df = pd.read_csv(path)  # use a consistent name
    df["source"] = path.split("/")[-1]  # add a 'source' column with the file name
    dfs.append(df)

# Merge all dataframes into one
data = pd.concat(dfs, ignore_index=True)
print("✅ Loaded data shape:", data.shape)
data.head()
```

```python
# Imports
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Suppose your merged CSV data is in 'data' with columns: 'PhD', 'Postdoc', 'source'
# Combine text from PhD and Postdoc
data["text"] = data["phd"].astype(str) + " " + data["postdoc"].astype(str)
data = data.dropna(subset=["text"])  # Drop empty rows

# Features and labels
X = data["text"]
y = data["source"]

# Convert text to TF-IDF features
vectorizer = TfidfVectorizer()
X_tfidf = vectorizer.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Initialize Multinomial Naive Bayes
model = MultinomialNB()

# Train
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Test with new input
test_input = ["PhD ExampleName Postdoc ExampleName2"]
test_tfidf = vectorizer.transform(test_input)
prediction = model.predict(test_tfidf)
print("Predicted CSV/source:", prediction)
```
